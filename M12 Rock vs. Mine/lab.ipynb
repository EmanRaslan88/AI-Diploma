```python
X=df.drop(60,axis=1)
y=df[60]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,random_state=1)
```

```python
def train(clf,X_train,X_test,y_train,y_test):
    clf.fit(X_train,y_train)
    y_pred_t=clf.predict(X_train)
    y_pred=clf.predict(X_test)
    print(accuracy_score(y_pred_t,y_train))
    print(classification_report(y_test,y_pred))
    sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
    plt.show()
```

```python
clf=RandomForestClassifier()
train(clf,X_train,X_test,y_train,y_test)
```

```python
clf=LogisticRegression()
train(clf,X_train,X_test,y_train,y_test)
```

```python
clf=GaussianNB()
train(clf,X_train,X_test,y_train,y_test)
```

```python
clf=KNeighborsClassifier()
train(clf,X_train,X_test,y_train,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
clf=GradientBoostingClassifier()
train(clf,X_train,X_test,y_train,y_test)
```

```python
clf=AdaBoostClassifier()
train(clf,X_train,X_test,y_train,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures()
from sklearn.pipeline import Pipeline

classifiers=[
    ('ada',AdaBoostClassifier()),
    ('knn',KNeighborsClassifier()),
    ('logi',LogisticRegression())
]
```

```python
scores=[]
for _,classifier in classifiers:
    print(classifier)
    clf=Pipeline([('poly',poly),('clf',classifier)])
    clf.fit(X_train,y_train)
    scores.append(clf.score(X_test,y_test))
```

```python
scores
sns.barplot(x=['ada','knn','log'],y=scores)
```

```python
clf=Pipeline([('poly',poly),('clf',KNeighborsClassifier())])
clf.fit(X_train,y_train)
print(clf.score(X_test,y_test))
```
